トータルステーションTrimble S6のシリアル出力をTera Termでログ出力したテキストファイルを用意する

```python
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
# 実行前にスペースを削除しておく
# CSVフォーマット:日付,時刻,水平角,鉛直角,射距離,水平距離,X座標,Y座標,Z座標,
df = pd.read_csv('C://home/workspace/teraterm-log/20200611.txt', sep=',', header=None, names=['yyyymmdd', 'hhnnss', 'lateral_angle', 'vertical_angle', 'slope_dist', 'lateral_dist', 'x', 'y', 'z', '_'])
df = df.drop('_', axis=1)
df = df.dropna()
df['timestamp'] = df['yyyymmdd'] + ' ' + df['hhnnss']
df = df.drop('yyyymmdd', axis=1)
df = df.drop('hhnnss', axis=1)
df = df.reset_index(drop=True)
df.head()
```

```python
# check
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_aspect('equal')
ax1.scatter(df.x, df.y, s=1, c='red', label='x vs y')
ax1.set_xlabel('x [m]')
ax1.set_ylabel('y [m]')
#comment = "a = %s [mm]\nb = %s [mm]\nd = %s [mm]\nr = %s [mm]" % (a, b, d, r)
#ax1.text(0, 0, comment, size=9, c="black")
#ax1.legend(bbox_to_anchor=(1, 1), loc='lower right', borderaxespad=1, fontsize=9)
plt.show()
```
