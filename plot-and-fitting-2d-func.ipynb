2変数関数の最小二乗法フィティング

```python
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
def func(x, y, C):
    u = np.power((x / y), 2) / np.sqrt(1 + np.power((x / y), 2))
    z = (u / C) * (1 - np.exp(-u / np.power(C, 2)))
    return z
```

```python
step = 0.5
variable_min = -10
variable_max = 10
variable_range = np.arange(variable_min, variable_max + step, step)
n = len(variable_range)
df = pd.DataFrame()
df['x'] = np.repeat(variable_range, n)
df['y'] = list(variable_range) * n
C = 1.0
df['z'] = func(df.x, df.y, C)
```

```python
df
```

```python
img = np.array(df.z).reshape(n, n)
plt.imshow(img)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(variable_range, variable_range, img)
plt.show()
```
